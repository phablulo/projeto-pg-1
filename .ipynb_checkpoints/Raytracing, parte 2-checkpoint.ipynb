Primeiro, vamos importar as bibliotecas necessárias.

````python
import numpy as np
from numpy.linalg import norm
from math import pi, tan, atan, sqrt, sin, cos, inf
from matplotlib.pyplot import imshow
%matplotlib inline
np.seterr(invalid='raise')
````

E algumas funções que serão úteis:

````python
def showImage(image):
  # matplotlib espera hxw
  imshow(image.transpose(1, 0, 2))
  
def minQuadraticSolution(a,b,c):
  delta = b*b - 4*a*c
  if delta < 0:
    return None
  if delta == 0:
    return -b/(2*a)
  delta = sqrt(delta)
  # só importam soluções que que estão
  # a frente de onde a pessoa está olhando
  # então retorna o menor positivo!
  t2 = (-b -delta)/(2*a)
  if t2 < 0:
    t1 = (-b + delta)/(2*a)
    if t1 < 0:
      return None
    return t1
  return t2

def normalize(vector):
  return vector / norm(vector)

def lerp(start, end, t):
  return (1 - t)*start + t*end

def rotation_matrix(axis, theta):
  # https://stackoverflow.com/a/6802723/8589297
  """
  Return the rotation matrix associated with counterclockwise rotation about
  the given axis by theta radians.
  """
  axis = np.asarray(axis)
  axis = axis / sqrt(np.dot(axis, axis))
  a = cos(theta / 2.0)
  b, c, d = -axis * sin(theta / 2.0)
  aa, bb, cc, dd = a * a, b * b, c * c, d * d
  bc, ad, ac, ab, bd, cd = b * c, a * d, a * c, a * b, b * d, c * d
  return np.array([[aa + bb - cc - dd, 2 * (bc + ad), 2 * (bd - ac)],
                   [2 * (bc - ad), aa + cc - bb - dd, 2 * (cd + ab)],
                   [2 * (bd + ac), 2 * (cd - ab), aa + dd - bb - cc]])

def rotate(vector, axis, angle):
  M = rotation_matrix(axis, angle)
  rotated = np.matmul(M, vector)
  # remove valores muito pequenos
  rotated[np.abs(rotated) < 1e-6] = 0
  return rotated
````

Agora, vamos criar classes. Uma irá codificar nossos raios do RayTracer. Cada raio possui duas propriedades: origem e direção.

````python
class Ray:
  def __init__(self, origin, direction):
    self.origin = np.array(origin)
    self.direction = np.array(direction)
  def getPointAt(self, t):
    return self.origin + self.direction*t
````

Devemos possuir materiais. Cada material possui as propriedades:

- cor, dada em rgb, ([0, 1], [0, 1], [0, 1])
- fator difuso, o quanto ele reflete luz difusa [0, 1]
- fator specular, o quanto ele é reflexivo [0, 1]
- fator transmissivo, o quanto a luz passa por ele [?, ?]
- fator de refração, o quanto ele refrata luz [?, ?]
- fator emissivo, (não sei) [?, ?]
- alpha, o quanto ele dispersa a luz difusa [0, $\infty$]

````python
class Material:
  # rgb = cor
  # kd = fator difuso
  # ks = fator specular
  # kt = fator transmissivo
  # kr = fator de refração
  # alpha = coeficiente de rugosidade
  def __init__(self, rgb, kd, ks, kt, kr, ke=1, alpha=40):
    self.rgb = np.array(rgb)
    self.kd = kd
    self.ks = ks
    self.kt = kt
    self.kr = kr
    self.ke = ke
    self.alpha = alpha
````

E, agora, classes para representar os objetos em nosso universo.

Para saber se um raio intersecta um objeto, chame um dos métodos:

```python3
object.getRayIntersectionPoint(ray) -> Point
object.getRayIntersectionDistance(ray) -> number
```

Se eles retornarem `None`, é porque não há interseção.

````python
class Object:
  def __init__(self, material):
    self.material = material
    
  def getRayIntersectionDistance(self, ray):
    raise NotImplemented("Não implementado")
    
  def getRayIntersectionPoint(self, ray):
    t = self.getRayIntersectionDistance(ray)
    if t is None:
      return None
    return ray.getPointAt(t)
  
  def getIntersectionNormal(self, intersectionPoint):
    raise NotImplemented("Não implementado")

class Sphere(Object):
  def __init__(self, material, r, xyz):
    super().__init__(material)
    self.r = r
    self.xyz = np.array(xyz)
  def getRayIntersectionDistance(self, ray):
    # solve the quadratic equation!
    tmp = ray.origin - self.xyz
    a = norm(ray.direction)**2
    b = 2*np.dot(tmp, ray.direction)
    c = norm(tmp)**2 - self.r**2
    t = minQuadraticSolution(a,b,c)
    return t
  def getIntersectionNormal(self, intersectionPoint):
    return normalize(intersectionPoint - self.xyz)
    
class Plane(Object):
  # point é um ponto que pertence ao plano
  # normal é um vetor normal ao plano
  def __init__(self, material, point, normal):
    super().__init__(material)
    self.point = np.array(point)
    self.normal = normalize(np.array(normal))
  def getRayIntersectionDistance(self, ray):
    # vamos ver se o vetor direção de ray
    # intersecta o plano em algum ponto p.
    # isso significa que (p - point) × normal = 0
    # onde × denota o dot-product.
    # Se a normal é dada por (a, b, c), p por (x,y,z)
    # e point por (x0, y0, z0), então:
    # a(x - x0) + b(y - y0) + c(z - z0) = 0
    # => ax + by + cz - (ax0 + by0 + cz0) = 0
    #    --normal×p--   --normal×point--
    # :: ax + by + cz - d = 0
    d = np.dot(self.normal, self.point)
    # Mas como encontrar p = (x, y, z) ?
    # ray.origin + ray.direction*t = p
    # assim, reduzimos o problema a encontrar t
    # se ray.origin = (o1, o2, o3)
    # e ray.direction = (d1, d2, d3), então:
    # o1 + d1t = x
    # o2 + d2t = y
    # o3 + d3t = z
    # => a(o1 + d1t) + b(o2 + d2t) + c(o3 + d3t) - d = 0
    # => t(ad1 + bd2 + cd3) + ao1 + bo2 + co3 - d = 0
    # => t(ad1 + bd2 + cd3) = d - (ao1 + bo2 + co3)
    #                         -------it's d--------
    d = d - np.dot(self.normal, ray.origin)
    # => t(ad1 + bd2 + cd3) = d
    # => t = d / (ad1 + bd2 + cd3)
    s = np.dot(self.normal, ray.direction)
    # * Se o denominador é zero, eles só intersectam
    # no infinito. Aqui, vamos assimir que não intersectam
    if s == 0:
      return None
    t = d / s
    # * Se t é negativo, então o raio intersecta o plano
    # atrás de onde o raio está apontando. Portando, vamos
    # assumir que eles não se tocam:
    if t < 0:
      return None
    return t
  def getIntersectionNormal(self, intersectionPoint):
    return self.normal
````

Também precisamos definir as luzes!

````python
class Light:
  def __init__(self, xyz, difuse, specular):
    self.xyz = np.array(xyz)
    
    if isinstance(difuse, (int, float)):
      self.difuse = np.array((difuse, difuse, difuse))
    else:
      self.difuse = np.array(difuse)
    if isinstance(specular, (int, float)):
      self.specular = np.array((specular,specular,specular))
    else:
      self.specular = np.array(specular)
````

Agora que temos que criar uma forma de visualizar os objetos. Precisamos de uma câmera!

Câmeras possuem as seguintes propriedades:

- Field of View, seu ângulo (em rad) de abertura horizontal
- Posição, seu ponto no espaço
- Ponto de mira, vetor que indica o PONTO para o qual ela "olha"
- orientação, indica pra onde aponta a parte de cima da câmera
- Distânca para o plano de imagem

Note que Field of View e Distância para o Plano de Imagem estão relacionados. Quanto maior a distância, menor o FoV. 

````python
class Camera:
  # xyz é a posição
  # target é o ponto de que a câmera vai "olhar"
  # fov é o Field of View
  # dist é a distância para o plano de imagem
  # up é o vetor orientação
  def __init__(self, xyz, target, up, fov=pi/2, dist=None):
    self.xyz = np.array(xyz)
    self.target = normalize(np.array(target) - self.xyz)
    self.up = normalize(np.array(up))
    self.ambientLight = np.array((.6, .6, .6))
    
    self.w = 300
    self.h = 190
    
    if fov is not None:
      if dist is not None:
        raise AttributeError("Espefique FoV ou Distância, mas não os dois")
      self.setFov(fov)
    elif dist is not None:
      self.setDist(dist)
    else:
      raise AttributeError("Especifique FoV ou Distância")

  def setFov(self, fov):
    self.fov = fov
    # ajusta distancia para corresponder
    self.dist = 1/tan(fov/2)
  def setDist(self, dist):
    self.dist = dist
    # ajusta fov para corresponder
    # http://paulbourke.net/miscellaneous/lens/
    self.fov = 2*atan(1/dist)
  def setDistance(self, dist):
    self.setDist(dist)
  def setTarget(self, target):
    self.target = normalize(np.array(target) - self.xyz)
  def setPosition(self, xyz):
    self.xyz = np.array(xyz)
  def setXYZ(self, xyz):
    self.setPosition(xyz)
  def setXyz(self, xyz):
    self.setPosition(xyz)
  def setUp(self, up):
    self.up = normalize(np.array(up))
  def setOrientation(self, up):
    self.setUp(up)
    
  def _getImagePlaneBounds(self):
    ratio = self.h / self.w
    w = 2 # sempre 2
    h = w * ratio
    # encontra o centro do plano de imagem
    center = self.xyz + self.dist * self.target
    # up é um vetor que indica a direção da parte de cima da câmera.
    # vamos usá-lo pra calcular o deslocamento vertical e horizontal
    # que o ponto central precisa sofrer:
    vert = self.up*(h/2)
    hori = rotate(self.up, self.target, -pi/2)*(w/2)
    
    # desloca o centro para calular as quinas
    tl = center + ( vert - hori)
    tr = center + ( vert + hori)
    bl = center + (-vert - hori)
    br = center + (-vert + hori)
    
    return (tl, tr, bl, br)
  
  def _getRayForPixel(self, x, y, bounds):
    xt = x / self.w
    yt = y / self.h
    
    top = lerp(bounds[0], bounds[1], xt)
    bottom = lerp(bounds[2], bounds[3], xt)
    point = lerp(top, bottom, yt)
    
    return Ray(point, point - self.xyz)
    
  def _bgColorForRay(self, ray):
    # projeta a direção do raio sobre a direção pra cima da câmera.
    # Quanto mais eles estiverem alinhados, mais azul será o resultado.
    # Como os dois vetores são unitários, podemos calcular facilmente:
    # cos(theta) = ray.direction × self.up, com × sendo o dot-product
    alignment = np.dot(normalize(ray.direction), self.up)
    alignment = max(0, min(1, alignment))
    subtractR = 0.35294117647
    subtractG = 0.20784313725
    r = 1 - subtractR * alignment
    g = 1 - subtractG * alignment
    return np.array([r, g, 1])
  
  def _getClosestObject(self, ray, objects):
    smaller = None
    for item in objects:
      t = item.getRayIntersectionDistance(ray)
      if t is None:
        continue
      if smaller is None or t < smaller[0]:
        smaller = (t, item)
    return smaller
  
  def _isShadowed(self, ray, objects, ignore):
    smaller = None
    for item in objects:
      if item is ignore:
        continue
      t = item.getRayIntersectionDistance(ray)
      if t is None:
        continue
      if smaller is None or t < smaller:
        smaller = t
    # como a direção do raio não é normalizada,
    # um t > 1 significa que o objeto tocado está DEPOIS da lâmpada.
    # um t < 0 significa que o objeto tocado está atrás do objeto atual
    return smaller is not None and smaller > 0 and smaller <= 1
    
  def _colorForRay(self, ray, objects, lights, depth=3, kr=1):
    item = self._getClosestObject(ray, objects)
    if item is None:
      return self._bgColorForRay(ray)
    t, item = item
    material = item.material
    intersection = ray.getPointAt(t)
    normal = item.getIntersectionNormal(intersection)
    eyeDirection = normalize(ray.origin - intersection) # ray.origin -> self.xyz
    
    # refração!
    refrated = False
    if material.kt > 0:
      rfactor = kr / material.kr
      c1 = np.dot(normal, normalize(ray.direction))
      tmp = 1 - (rfactor**2)*(1 - c1**2)
      if tmp > 0:
        # refrata!
        c2 = sqrt(tmp)
        # T é o vetor transmissão
        # i.e a direção do raio refratado
        T = rfactor*ray.direction + (rfactor*c1 - c2)*normal
        rRay = Ray(
          intersection + T*0.01,
          T
        )
        newKr = 1 if kr != 1 else material.kr
        color = self._colorForRay(rRay, objects, lights, kr=newKr) * material.kt
        refrated = True
      else:
        # não refrata
        pass
    
    # se o material refratou, então eu vou ignorar
    # a cor do material dele. Sua cor será apenas
    # o resultado da refração + as reflexões
    if not refrated:
      color = material.rgb * self.ambientLight
    
    # iluminação
    for light in lights:
      lightDirection = normalize(light.xyz - intersection)
      isShadow = self._isShadowed(
        Ray(intersection, lightDirection),
        objects,
        item
      )
      if isShadow:
        # está na sombra
        continue
      lightNormal = np.dot(normal, lightDirection)
      if lightNormal < 0:
        # luz está vindo por dentro do objeto
        continue
      # luz difusa
      difuse = material.kd * light.difuse * lightNormal
      # luz especular
      reflectance = 2*normal*lightNormal - lightDirection
      specular = material.ks * light.specular
      specular = specular * np.dot(eyeDirection, reflectance)**material.alpha
      # junta tudo
      color = color + difuse + specular
    # reflexão
    if not refrated and depth > 0:
      rd = normalize(-ray.direction)
      reflectance = 2*normal*np.dot(rd, normal) - rd
      reflectionRay = Ray(
        intersection + reflectance*0.01,
        reflectance
      )
      reflected = self._colorForRay(reflectionRay, objects, lights, depth-1)
      color = color + reflected * material.ks
    
    return np.clip(color, 0, 1)
    
  def showDirections(self):
    bounds = self._getImagePlaneBounds()
    r = np.zeros((self.w, self.h))
    g = np.zeros((self.w, self.h))
    for x in range(self.w):
      for y in range(self.h):
        ray = self._getRayForPixel(x, y, bounds)
        r[x, y] = ray.direction[0]
        g[x, y] = ray.direction[1]
    colors = np.zeros((self.w, self.h, 3))
    def normalizeColor(color):
      color = color - np.amin(color) # get length
      color = color / np.amax(color) # get percent
      return color
    colors[:,:,0] = normalizeColor(r)
    colors[:,:,1] = normalizeColor(g)
    showImage(colors)
    
  def takePicture(self, objects, lights):
    bounds = self._getImagePlaneBounds()
    colors = np.zeros((self.w, self.h, 3))
    for x in range(self.w):
      for y in range(self.h):
        ray = self._getRayForPixel(x, y, bounds)
        color = self._colorForRay(ray, objects, lights)
        colors[x, y] = color
    showImage(colors)
````

````python
green = Material(rgb=(.1,.9,.3), kd=.4, ks=.2, kt=.0, kr=.0)
blue  = Material(rgb=(.3,.2,.9), kd=.4, ks=.0, kt=.0, kr=.0)
red   = Material(rgb=(.8,.3,.2), kd=.5, ks=.4, kt=.0, kr=.0)
gray  = Material(rgb=(.8,.8,.8), kd=.5, ks=.3, kt=.9, kr=1.7)

objects = (
  Plane(green, point=(0, -1, 0), normal=(0, 1, 0)),
  Sphere(red,  r=1., xyz=(-0.8, 0.0, 4)),
  Sphere(blue, r=.6, xyz=(-2.5, -.4, 3)),
  Sphere(gray, r=4., xyz=( 3.0, 3.0, 6)),
)

lights = (
  Light(xyz=(3, 2, 3), difuse=.4, specular=.4),
  Light(xyz=(-3, 2, 2), difuse=.4, specular=.4)
)

camera.setFov(pi/2)
camera.takePicture(objects, lights)
````
